```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = './books.csv'  
books_df = pd.read_csv(file_path)
```

```python
# Q1: Number of rows and columns in the dataset
num_rows, num_cols = books_df.shape
num_rows, num_cols
```

```python
# Q2: Number of books without an original title
books_no_original_title = books_df['original_title'].isna().sum()
books_no_original_title
```

```python
# Q3: Number of unique books in the dataset after removing null values in the original_title column
unique_books = books_df[books_df['original_title'].notna()]['book_id'].nunique()
unique_books
```

```python
# Q4: Average rating of all books in the dataset
average_rating = books_df['average_rating'].mean()
average_rating
```

```python
# Q5: Number of books published in the year 2000
books_2000 = books_df[books_df['original_publication_year'] == 2000].shape[0]
books_2000
```

```python
# Q6: Book title with the maximum number of ratings
max_ratings_book = books_df[books_df['work_ratings_count'] == books_df['work_ratings_count'].max()]['original_title'].iloc[0]
max_ratings_book
```

```python
# Q7: Bucket the average_rating of books into 11 buckets [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0] with 0.5 decimal rounding (eg: average_rating 3.0 to 3.49 will fall in bucket 3.0). Plot bar graph to show total number of books in each rating bucket.
rating_buckets = books_df['average_rating'].apply(lambda x: round(x * 2) / 2)
bucket_counts = rating_buckets.value_counts().sort_index()
bucket_counts
```

```python
plt.figure(figsize=(10, 6))
bucket_counts.plot(kind='bar')
plt.title('Number of Books in Each Rating Bucket')
plt.xlabel('Average Rating Buckets')
plt.ylabel('Number of Books')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
